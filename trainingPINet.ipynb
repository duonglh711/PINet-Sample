```python
#setup data
import os
os.chdir(data_folder)
import glob2
import math  
import numpy as np
 
files = []
for ext in ["*.png", "*.jpeg", "*.jpg"]:
  image_files = glob2.glob(ext)
  files += image_files
    

nb_val = math.floor(len(files)*0.2)
rand_idx = np.random.randint(0, len(files), nb_val)

# Tạo file train.txt
with open(os.path.join(root_folder, "train.txt"), "w+") as f:
  for idx in np.arange(len(files)):
    text_file = files[idx][:-3] + "txt"
    if os.path.exists(text_file) and idx not in rand_idx:
      f.write(os.path.join(data_folder,files[idx])+'\n')

        
# Tạo file val.txt
with open(os.path.join(root_folder, "val.txt"), "w+") as f:
  for idx in np.arange(len(files)):
    text_file = files[idx][:-3] + "txt"
    if os.path.exists(text_file) and idx in rand_idx:
      f.write(os.path.join(data_folder,files[idx])+'\n')
      

print(f'train: {len(files)-nb_val}, val: {nb_val}')
```
